```python
consecutive_sequences = {
        '00': 0,
        '01': 0,
        '10': 0,
        '11': 0
    }


def find_consecutive_sequences(seq):
   
    
    current_sequence = ''
    
    for bit in seq:
        current_sequence += str(bit)
        
        # Check if the current sequence has reached a length of 2
        if len(current_sequence) == 2:
            # Update the count for the current sequence
            consecutive_sequences[current_sequence] += 1
            
            # Reset the current sequence to the last bit
            current_sequence = current_sequence[-1]
    
    return consecutive_sequences


consecutive_sequences = find_consecutive_sequences(XXY)


# Print the counts of consecutive sequences
print("Counts of Consecutive Sequences:")
print("00:", consecutive_sequences['00'])
print("01:", consecutive_sequences['01'])
print("10:", consecutive_sequences['10'])
print("11:", consecutive_sequences['11'])



n01 = consecutive_sequences['01']
n11 = consecutive_sequences['11']
n10 = consecutive_sequences['10']
n00 = consecutive_sequences['00']



n = n00+n01+n10+n11

print(n)



p00 = n00/ (n01+n00)

p01 = n01 / (n01+n00)

p11 = n11 / (n10+n11)

p10 = n10 /(n10+n11)


print(p00,p01,p10,p11)



n0_ = n00+n01
n_0 = n10 + n00
n1_ = n10+n11
n_1 = n01+n11


e00 = n0_*n_0/n
e01 = n0_ * n_1/n
e10 = n1_* n_0/n
e11 = n1_ * n_1/n

print('expected')
print(e00,e01,e10,e11)
```
