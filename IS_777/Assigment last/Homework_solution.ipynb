```python
import random
import numpy as np 
x1=[]
for i in range(100000):
    x = random.random()
    x1.append((x))
    
#print(x1)    
    #
x1=np.array(x1)
x1[np.where(x1 <0.5)]=-1 
x1[np.where(x1 >0.5)]=1 
#x1
#test=x1

#print(test) 
x11=[]
for i in range(100000):
    x = random.random()
    x11.append((np.round(x)))
    #
x11=np.array(x11)
x11[np.where(x11 <0.1)]=-1 
x11[np.where(x11 >0.1)]=1 



x111=[]
for i in range(100000):
    x = random.random()
    x111.append((np.round(x)))
    #
x111=np.array(x111)
x111[np.where(x111 <0.9)]=-1 
x111[np.where(x111 >0.9)]=1 



#x1
###testkit 1 results
testkit1=np.concatenate((x1,x11),axis=0)

###testkit 2 results
testkit2=np.concatenate((x11,x11),axis=0)


#actual diagnosis
fact=np.concatenate((x1,x111),axis=0)
np.shape(fact)

```

```python
#Q1. Which test kit is better? and why? 50

#I will choose the testkit1 as it shows better accuracy than testkit2


def geting_TPR_FPR(test):

    TP=[];TN=[];FP=[];FN=[]
    for i in range(len(test)):
        if test[i]==1 and fact[i] ==1:
            TP.append(1)
        if test[i]==-1 and fact[i] ==-1:
            TN.append(1)      
        if test[i]==1 and fact[i] ==-1:
            FP.append(1)  
        if test[i]==-1 and fact[i] ==1:
            FN.append(1)   

    TP=np.sum(TP)
    FP=np.sum(FP)
    TN=np.sum(TN)
    FN=np.sum(FN)

    TPR = TP / (TP + FN)
    FPR = FP / (FP + TN)
    print(FPR)

    FNR=1-TPR
    TNR=1-FPR

    return TPR,FPR,FNR,TNR

TPR,FPR,FNR,TNR = geting_TPR_FPR(testkit1)

print('\n')
print('********************************************************')
print('Analysis for testkit1')
print('********************************************************')
P_AbyN=2/1000
print(P_AbyN)
#P(Person tested has COVID ∣ TEST1 is positive) = P(Person tested has COVID & TEST1 is positive)/P(TEST1 is positive)
#NUM
P_E1_and_A=P_AbyN * TPR #0.00148 * 0.93 
print(P_E1_and_A)

#DENOM
#P_E1=P_E1_and_A+P_E1_and_Aneg
P_E1_and_Aneg=(1-P_AbyN)*FPR
P_E1=P_E1_and_A+P_E1_and_Aneg
print(P_E1)
P_acc=P_E1_and_A/P_E1
print('accuracy')
print(P_acc*100)



print('\n')
print('********************************************************')
print('Analysis for testkit2')
print('********************************************************')



TPR,FPR,FNR,TNR = geting_TPR_FPR(testkit2)



P_AbyN=2/1000
print(P_AbyN)
#P(Person tested has COVID ∣ TEST1 is positive) = P(Person tested has COVID & TEST1 is positive)/P(TEST1 is positive)
#NUM
P_E1_and_A=P_AbyN * TPR #0.00148 * 0.93 
print(P_E1_and_A)

#DENOM
#P_E1=P_E1_and_A+P_E1_and_Aneg
P_E1_and_Aneg=(1-P_AbyN)*FPR
P_E1=P_E1_and_A+P_E1_and_Aneg
print(P_E1)
P_acc=P_E1_and_A/P_E1
print('accuracy')
print(P_acc*100)



#Q1. Which test kit is better? and why? 50

#I will choose the testkit1 as it shows better accuracy than testkit2
```
