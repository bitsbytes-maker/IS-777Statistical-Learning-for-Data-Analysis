```python
# rice production data 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os

data = pd.read_csv('rice production across different countries from 1961 to 2021.csv')

data_fran = data[data['Area'] == 'France']['Value']

print(len(data_fran))

data_usa = data[data['Area'] == 'United States of America']['Value']

print(len(data_usa))

data_bang = data[data['Area'] == 'Bangladesh']['Value']

print(len(data_bang))



def standardscaling(data_bang):

    data_=(data_bang-np.mean(data_bang))/np.std(data_bang)

    return data_

data_bang = standardscaling(data_bang)
data_fran = standardscaling(data_fran)
data_usa = standardscaling(data_usa)
```

```python
import numpy as np
from numpy.linalg import eig


def eigen_value_vector(data_usa,data_bang):
    x = np.array([data_usa,data_bang])

    a = np.cov(x)
    w,v=eig(a)

    print('E-value:', w)
    print('E-vector', v)

    return w,v


print('USA and Bangladesh')
w_usa_bang, v_usa_bang = eigen_value_vector(data_usa,data_bang)
print('USA and France')
w_usa_fran, v_usa_fran = eigen_value_vector(data_usa,data_fran)
print('Bangladesh and France')
w_bang_fran, v_bang_fran = eigen_value_vector(data_bang,data_fran)
```

```python
import numpy as np

import math 


print('USA and Bangladesh')
print('First eigen vectors angle with x axis')
print(np.arctan(-1*v_usa_bang[0][1]/ v_usa_bang[0][0])*180/3.14)
print('Second eigen vectors angle with x axis')
print(np.arctan(-1*v_usa_bang[1][1]/ v_usa_bang[1][0])*180/3.14)

# Calculate the eigenvectors and eigenvalues
eigenvalues, eigenvectors =w_usa_bang, v_usa_bang 

# Sort the eigenvectors based on eigenvalues (descending order)
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Choose two eigenvectors for which you want to find the angle
eigenvector1 = sorted_eigenvectors[:, 0]
eigenvector2 = sorted_eigenvectors[:, 1]

# Calculate the cosine of the angle between the eigenvectors
cosine_angle = np.dot(eigenvector1, eigenvector2) / (np.linalg.norm(eigenvector1) * np.linalg.norm(eigenvector2))

# Calculate the angle in radians
angle_radians = np.arccos(cosine_angle)

# Convert the angle to degrees
angle_degrees = np.degrees(angle_radians)

# Display the result
print('USA and Bangladesh')
print(f"Angle between the first and second eigenvectors: {angle_degrees} degrees.")
```

```python





def returnpc1_pc2(v_bang_fran,data_bang,data_fran):
    PC1=[]
    PC2=[]


    for i in range(61):
        PC1.append(v_bang_fran[0][0]*np.array(data_bang)[i] + v_bang_fran[0][1]*np.array(data_fran)[i])
        PC2.append(v_bang_fran[1][0]*np.array(data_bang)[i] + v_bang_fran[1][1]*np.array(data_fran)[i])
    
    return PC1,PC2


```

```python
PC1,PC2 = returnpc1_pc2(v_usa_bang,data_usa,data_bang)


timeseries = np.linspace(1,61,61)

plt.plot(timeseries, np.array(data_usa), label='USA', color='blue')
plt.plot(timeseries, np.array(data_bang), label='Bangladesh', color='green')
plt.plot(timeseries, PC1, label='PC1', alpha=0.5, linestyle='dashed', color='red')
plt.plot(timeseries, PC2, label='PC2', alpha=0.5, linestyle='dashed', color='orange')

plt.legend()
plt.show()
```
