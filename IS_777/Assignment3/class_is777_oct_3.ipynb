```python
consecutive_sequences = {
        '00': 0,
        '01': 0,
        '10': 0,
        '11': 0
    }


def find_consecutive_sequences(seq):
   
    
    current_sequence = ''
    
    for bit in seq:
        current_sequence += str(bit)
        
        # Check if the current sequence has reached a length of 2
        if len(current_sequence) == 2:
            # Update the count for the current sequence
            consecutive_sequences[current_sequence] += 1
            
            # Reset the current sequence to the last bit
            current_sequence = current_sequence[-1]
    
    return consecutive_sequences
```

```python


import pandas as pd
import re

# Load the Excel file into a DataFrame
df = pd.read_excel('Exercise_covidyear.xlsx')

print(df.columns)


column_name = 'exercise'

total_0_1_val = df[column_name].values


consecutive_sequences = {
        '00': 0,
        '01': 0,
        '10': 0,
        '11': 0
    }

def find_consecutive_sequences(seq):
   
    
    current_sequence = ''
    
    for bit in seq:
        current_sequence += str(bit)
        
        # Check if the current sequence has reached a length of 2
        if len(current_sequence) == 2:
            # Update the count for the current sequence
            consecutive_sequences[current_sequence] += 1
            
            # Reset the current sequence to the last bit
            current_sequence = current_sequence[-1]
    
    return consecutive_sequences

# Example list of 0s and 1s
sequence_list = total_0_1_val

# Find consecutive sequences in the list
consecutive_sequences = find_consecutive_sequences(sequence_list)

print(total_0_1_val)

# Print the counts of consecutive sequences
print('observed')
print("Counts of Consecutive Sequences:")
print("00:", consecutive_sequences['00'])
print("01:", consecutive_sequences['01'])
print("10:", consecutive_sequences['10'])
print("11:", consecutive_sequences['11'])

n01 = consecutive_sequences['01']
n11 = consecutive_sequences['11']
n10 = consecutive_sequences['10']
n00 = consecutive_sequences['00']


```

```python

import numpy as np
from scipy.stats import chisquare
obs = [n00,n01,n10,n11]

eop = [e00,e01,e10,e11]


print(obs)
print(eop)




# Perform the chi-squared test
chi_squared, p_value = chisquare(f_obs=obs, f_exp=eop)

# Print the results
print(f"Chi-squared statistic: {chi_squared}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Observed and expected values are significantly different.")
else:
    print("Observed and expected values are not significantly different.")
```
